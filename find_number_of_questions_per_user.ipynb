# find_number_of_questions_per_user.ipynb

This file contains the code for finding the number of questions answered by each user, useful for generating the figures in the paper.

```python
# Import libraries

import pandas as pd
```

```python
# Read raw data
transactions_df = pd.read_csv('dataverse_files/2_DBE_KT22_datafiles_100102_csv/Transaction.csv')

# Check if it is read correctly
transactions_df.head()
```

```python
student_ids = transactions_df['student_id'].unique()
student_dict = {} # a dictionary with keys of student_id and values of the number of rows for each student_id
for student_id in student_ids:
    student_dict[student_id] = transactions_df[transactions_df['student_id'] == student_id].shape[0]

# Converting to a dataframe
student_df = pd.DataFrame.from_dict(student_dict, orient='index')
student_df.reset_index(inplace=True)
student_df.columns = ['student_id', 'num_questions']

# Check the results
student_df.head()
```

```python
student_df.to_csv('number_of_questions_per_student.csv', index=False) # Saving to CSV for plot generation later on
```
